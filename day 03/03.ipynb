```python
f = open("./input", "r")
c = f.read()
f.close()

lines = c.split("\n")
lines.pop()

down = 1
right = 3

curx = 0
cury = 0
width = len(lines[0]) - 1
trees = 0

for line in lines:        
    if curx > width:
        curx -= (width+1)
    if(line[curx] == "#"):
        print ("oh a tree at ", cury, curx)
        trees += 1
    curx += right
    cury += down
    

print("got ", trees, "trees in the way.")
```
