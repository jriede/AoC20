```python
import re
f = open("./input", "r")
c = f.read()
f.close()

lines = c.split("\n")
lines.pop()

pwds = []

for l in lines:
    line = l.split(" ")
    pwds.append(line)

print(len(pwds))
```

```python
def find_valid_pwds(pwds):
    # number of valid passwords
    valid = 0
    for p in pwds:
        #print(p)
        #m = re.match(r"(\d+)-(\d+)(\w+)[a-z+]: (\w+)", p)
        m = re.match(r"(\d+)-(\d+) ([a-z+]): (\w+)", p)
        #print(m.group(0))
        r1 = int(m.group(1)) # min
        r2 = int(m.group(2)) # max
        letter = m.group(3)
        pw = m.group(4)
        print(r1, r2, letter, pw)
        whitelist = set(letter)
        pw = ''.join([c for c in pw if c in whitelist])
        print(r1, r2, letter, pw)
        c = 0
        for i in pw: 
            if i == letter: 
                c = c + 1
        if c >= r1 and c <= r2:
            print("pw valid (", c, ")")
            valid = valid + 1
        else:
            print("pw not valid (", c, ")")
        
    return valid
```

```python
val = find_valid_pwds(lines)
```

```python
# part two
#same, but with positions r1 and r2 instead of a range
def find_valid_pwds_2(pwds):
    # number of valid passwords
    valid = 0
    for p in pwds:
        m = re.match(r"(\d+)-(\d+) ([a-z+]): (\w+)", p)
        #print(m.group(0))
        r1 = int(m.group(1)) # position 1
        r2 = int(m.group(2)) # position 2
        letter = m.group(3)
        pw = m.group(4)
        print(r1, r2, letter, pw, pw[0])

        if pw[r1-1] == letter and pw[r2-1] != letter:
            print("pw valid (", letter, " at pos. ", str(r1), " but not at pos. ", r2, ")")
            valid = valid + 1
        elif  pw[r1-1] != letter and pw[r2-1] == letter:
            print("pw valid (", letter, " not at pos. ", str(r1), " but at pos. ", r2, ")")
            valid = valid + 1
        else:
            print("pw not valid")
        
    return valid
```

```python
val = find_valid_pwds_2(lines)
```
