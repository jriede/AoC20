```python
f = open("./input", "r")
c = f.read()
f.close()
lines = c.split("\n")
lines.pop()
top = 0

seats = []

for line in lines:
    ul = 127
    ll = 0
    for i in range(7):
        if line[i] == "F":
            ul -= (ul-ll+1)/2
            #print('F', ll, ul)
        elif line[i] == "B":
            ll += (ul-ll+1)/2
            #print('B', ll, ul)
    row = int(ll)
    #print ("row ", str(row))
    
    ul = 7
    ll = 0
    for i in range(7,10):
        
        if line[i] == "L": # keep lower half
            ul -= (ul-ll+1)/2
            #print('L', ll, ul)
        elif line[i] == "R": # keep upper half
            ll += (ul-ll+1)/2
            #print('R', ll, ul)
        #print(line[i])
    col = int(ll)
    #print ("col ", str(col))

    #print(row*8 + seat)
    seats.append([row,col])
    if row*8 + seat > top:
        top = row*8 + seat
print(top)
```

```python
# part two
occ = [[' ' for col in range(8)] for row in range(128)]
for seat in seats:
    occ[seat[0]][seat[1]] = 'X'
    
for i in range(len(occ)):
    s = str(i)
    for j in range(len(occ[i])):
        s = s + occ[i][j]
        #print(i,j, occ[i][j])
    print(s)
```
